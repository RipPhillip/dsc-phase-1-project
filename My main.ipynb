```python
# Turn columns into integers without $ signs
def convert_col_to_int(df, col):
    df[col] = df[col].str.replace("$", "").str.replace(",", "").astype('int')
    return df
```

```python
cols_to_convert = ['production_budget', 'domestic_gross', 'worldwide_gross']

for col in cols_to_convert:
     tn_budget_df = convert_col_to_int(tn_budget_df, col)
```

```python
with_studio_budget_df = pd.merge(movie_gross_df, tn_budget_df, how='left',
                                 left_on = ['title'], right_on=['movie'])
```

```python
with_studio_budget_df.drop(['id', 'release_date', 'movie', 'production_budget',
                           'domestic_gross_y', 'worldwide_gross'], axis=1, inplace=True)
```

```python
# Dropping NaN values
with_studio_budget_df.dropna(subset=['foreign_gross'],inplace=True)
```

```python
with_studio_budget_df['foreign_gross'] = with_studio_budget_df.iloc[:,3].apply(pd.to_numeric,errors='coerce')
```

```python
sequel_or_uni = ['Star Wars: The Force Awakens', 'Black Panther', 'Avengers: Infinity War',
                'Jurassic World', 'Marvel\'s The Avengers', 'Star Wars: The Last Jedi',
                'Incredibles 2', 'Rogue One: A Star Wars Story', 'Finding Dory'
                'Avengers: Age of Ultron','The Dark Knight Rises', 'The Hunger Games: Catching Fire',
                'Jurassic World: Fallen Kingdom', 'Toy Story 3', 'Wonder Woman',
                'Iron Man 3', 'Captain America: Civil War', 'Guardians of the Galaxy Vol. 2	',
                'Harry Potter and the Deathly Hallows Part 2', 'Despicable Me 2']
non_sequel = ['Beauty and the Beast (2017)', 'Frozen','The Secret Life of Pets', 'The Hunger Games',
             'Jumanji: Welcome to the Jungle']

```

```python
#Adding in production budget

with_studio_budget_df['production_budget'] = tn_budget_df['production_budget']
```

```python
# Creating a worldwide gross

with_studio_budget_df['worldwide_gross'] = with_studio_budget_df['domestic_gross_x'] +\
                                        with_studio_budget_df['foreign_gross']
```

```python
# Creating an ROI column
with_studio_budget_df['roi']=with_studio_budget_df['worldwide_gross'] - \
                            with_studio_budget_df['production_budget']
```

```python
# Creating a Star Wars column
tn_budget_df['Star_wars'] = tn_budget_df['movie']\
.map(lambda x: True if x in star_wars_movies else False)
```
