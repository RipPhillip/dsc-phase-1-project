# Marvel movies, a timeline

```python
def convert_col_to_int(df, col):
    df[col] = df[col].str.replace("$", "").str.replace(",", "").astype('int')
    return df
```

```python
cols_to_convert = ['production_budget', 'domestic_gross', 'worldwide_gross']

for col in cols_to_convert:
     tn_budget_df = convert_col_to_int(tn_budget_df, col)
        
```

```python
marvel_movies = ['Iron Man', 'The Incredible Hulk', 'Iron Man 2', 'Thor',
                'Captain America: The First Avenger', 'Marvel\'s The Avengers',
                'Iron Man 3', 'Thor: The Dark World', 'Captain America: The Winter Soldier',
                'Guardians of the Galaxy', 'Ant-Man', 'Avengers: Age of Ultron',
                'Captain America: Civil War', 'Doctor Strange', 
                'Guardians of the Galaxy Vol. 2', 'Spider-Man: Homecoming',
                'Thor: Ragnarok', 'Black Panther', 'Avengers: Infinity War',
                'Ant-Man and the Wasp', 'Captain Marvel', 'Avengers: Endgame',
                'Spider-Man: Far From Home']

len(marvel_movies)
```

```python
with_studio_budget_df = pd.merge(movie_gross_df, tn_budget_df, how='left',
                                 left_on = ['title'], right_on=['movie'])
```

```python
with_studio_budget_df.drop(['id', 'release_date', 'movie', 'production_budget',
                           'domestic_gross_y', 'worldwide_gross'], axis=1, inplace=True)
```

```python
# Dropping NaN values
with_studio_budget_df.dropna(subset=['foreign_gross'],inplace=True)
```

```python
with_studio_budget_df['foreign_gross'] = with_studio_budget_df.iloc[:,3].apply(pd.to_numeric,errors='coerce')
```

```python
with_studio_budget_df.info()
```

```python
with_studio_budget_df['Marvel'] = with_studio_budget_df['title']\
.map(lambda x: True if x in marvel_movies else False)
```

```python
sns.catplot(x='Marvel', y='domestic_gross_x', data=with_studio_budget_df)
```

```python
sns.barplot(x='year', y='Marvel', 
            data=with_studio_budget_df.sort_values(by='domestic_gross_x',
                                                  ascending=False))
```
